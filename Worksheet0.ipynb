```python
import matplotlib.pyplot as plt

time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]


study_hours = []
study_minutes = []

entertainment_hours = []

sleep_hours = []

low_study = []
moderate_study = []
high_study = []

for record in time_data:
  study_hours.append(record[0])

  entertainment_hours.append(record[1])

  sleep_hours.append(record[2])

  if record[0] < 3:
    low_study.append(record[0])
  elif 3 <= record[0] <= 5:
    moderate_study.append(record[0])
  else:
    high_study.append(record[0])

for hour in study_hours:
  study_minutes.append(hour * 60)


# Scatter plot
plt.scatter(study_hours, sleep_hours, color='blue', marker='o')
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Visualization: Study vs Sleep Pattern")
plt.show()


# Averages of Study, Entertainment and Sleep
average_study = sum(study_hours) / len(study_hours)

average_entertainment = sum(entertainment_hours) / len(entertainment_hours)

average_sleep = sum(sleep_hours) / len(sleep_hours)



print("\nStudy Hours:", study_hours)
print("------------------------------------------")
print("Study Minutes:", study_minutes)
print("------------------------------------------")


print("Low Study:", low_study)
print("Number of low study days:", len(low_study))
print("------------------------------------------")
print("Moderate Study:", moderate_study)
print("Number of moderate study days:", len(moderate_study))
print("------------------------------------------")
print("High Study:", high_study)
print("Number of high study days:", len(high_study))
print("------------------------------------------")


print("Average hours spent studying:", average_study)
print("------------------------------------------")
print("Average hours spent on entertainment:", average_entertainment)
print("------------------------------------------")
print("Average hours spent sleeping :", average_sleep)
```

```python
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]

def sum_nested_list(nested_list):
  """
  Calculates the sum of all numbers in a nested list.
  The function takes a list that may contain int or other nested lists as it's element.
  Args:
  nested_list
  Returns:
  total(sum)
  """

  total = 0
  for element in nested_list:
    if isinstance(element, list): # Check if the element is a list
      total += sum_nested_list(element) # Recursively sum the nested list
    else:
      total += element # Add the number to the total
  return total

print("The sum of given list is:", sum_nested_list(nested_list))
```

```python
def generate_permutations(s):
    """
    Generate all unique permutations of a given string using recursion.

    Parameters:
        s (str): The input string whose permutation is required.

    Returns:
        list: A list containing all unique permutations of the string.
    """
    # Base case: if the string has length 1, return it as the only permutation
    if len(s) == 1:
        return [s]

    permutations = []

    # Iterate through each character in the string
    for i, char in enumerate(s):
        # Remaining string after removing the current character
        remaining = s[:i] + s[i+1:]

        # Recursively generate permutations of the remaining string
        for permut in generate_permutations(remaining):
            new_permut = char + permut
            if new_permut not in permutations:
                permutations.append(new_permut)

    return permutations


print(generate_permutations("abc"))

print(generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    """
    Recursively calculate the total size of a directory.

    Parameters:
        directory (dict): A dictionary where keys are file names (values are sizes in KB)
                          or directory names (values are another dictionary representing a subdirectory).

    Returns:
        int: The total size of the directory in KB, including all nested subdirectories.
    """
    total_size = 0

    for name, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value

    return total_size


# Sample directory
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print("Total directory size:", calculate_directory_size(directory_structure), "KB")
```

```python
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount.
    """

    dp = [float('inf')] * (amount + 1)

    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)


    return dp[amount] if dp[amount] != float('inf') else -1


print(min_coins([1, 2, 5], 11))
print(min_coins([2], 3))
```

```python
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the longest common subsequence between two strings.
    """
    n, m = len(s1), len(s2)

    dp = [[0] * (m + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[n][m]


print(longest_common_subsequence("abcde", "ace"))
print(longest_common_subsequence("abc", "abc"))
print(longest_common_subsequence("abc", "def"))
```

```python
def knapsack(weights, values, capacity):
    """
    Finds the maximum value achievable within the given weight capacity.
    """
    n = len(weights)

    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                take = values[i - 1] + dp[i - 1][w - weights[i - 1]]
                skip = dp[i - 1][w]
                dp[i][w] = max(take, skip)
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]


weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))
```
