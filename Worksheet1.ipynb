# Matrix Operation


```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

1.	Prove A.A−1 = I.

```python
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)
print("A * A^-1:\n", identity)
```

2.	Prove AB ̸= BA.

```python
AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB:\n", AB)
print("BA:\n", BA)
print("AB == BA ?", np.array_equal(AB, BA))
```

3.	Prove (AB)T: = BT*AT .

```python
lhs = (AB).T
rhs = np.dot(B.T, A.T)
print("(AB)^T:\n", lhs)
print("B^T A^T:\n", rhs)
print("LHS and RHS Equal?", np.allclose(lhs, rhs))
```

•	Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1

x − y + 2z = −3

3x + y − z = 9


```python
# Coefficient matrix
coeff = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3,  1, -1]])

const = np.array([-1, -3, 9])

solution = np.linalg.solve(coeff, const)
print("Solution (x, y, z):", solution)

# Solving using inverse method explicitly
coeff_inv = np.linalg.inv(coeff)
solution_inv = np.dot(coeff_inv, const)
print("Solution via inverse:", solution_inv)
```
