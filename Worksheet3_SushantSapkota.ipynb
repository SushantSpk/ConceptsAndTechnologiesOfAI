# 3.2 Exploratory Data Analysis Practice Exercise - 1.

Which passenger had the highest fare paid relative to their age?

```python
t = pd.read_csv("/content/drive/MyDrive/Concept and Technology of AI/Titanic-Dataset.csv")

# Filling missing ages with median
age_median = t['Age'].median()
t['Age_filled'] = t['Age'].fillna(age_median)


t['fare_per_year'] = t['Fare'] / t['Age_filled']


high_fare_age = t[t['fare_per_year'] > 5].copy()


high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)


result = high_fare_age_srt[['Name', 'fare_per_year']].copy()

print("Passengers with high fare per age, sorted by fare_per_year:")
display(result.head(20))
```

Which adult male passenger (age â‰¥ 18 and Sex is â€™maleâ€™) paid the highest fare relative to their
class?

```python
t = pd.read_csv("/content/drive/MyDrive/Concept and Technology of AI/Titanic-Dataset.csv")

# Filling missing Age with median for safe comparisons
age_median = t['Age'].median()
t['Age'] = t['Age'].fillna(age_median)


t['fare_per_class'] = t['Fare'] / t['Pclass']


adult_males = t[(t['Sex'] == 'male') & (t['Age'] >= 18)].copy()


adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)


result = adult_males_srt[['Name', 'Age', 'fare_per_class']].copy()

print("Top adult male passengers by fare_per_class:")
display(result.head(20))
```

# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

What percent of the total fare revenue came from each passenger class?

```python
t = pd.read_csv("/content/drive/MyDrive/Concept and Technology of AI/Titanic-Dataset.csv")


total_fare = t['Fare'].sum()


fare_pclass = t.groupby('Pclass')['Fare'].sum().sort_index()


totals_list = [fare_pclass.get(1, 0.0), fare_pclass.get(2, 0.0), fare_pclass.get(3, 0.0)]


proportions = [(x / total_fare) * 100 for x in totals_list]

# Displaying nicely
fare_by_class_df = pd.DataFrame({
    'Pclass': [1,2,3],
    'FareTotal': totals_list,
    'PctOfTotalFare': proportions
})
fare_by_class_df['PctOfTotalFare'] = fare_by_class_df['PctOfTotalFare'].round(2)

print("Total fare and percentage of fare revenue by class:")
display(fare_by_class_df)
```

What percent of the total number of passengers on the Titanic belonged to each age group
(e.g., child, adult, senior)?

```python
t = pd.read_csv("/content/drive/MyDrive/Concept and Technology of AI/Titanic-Dataset.csv")

# For grouping, what to do with missing Age: dropping those rows (as they cannot be categorized)
t_age = t.dropna(subset=['Age']).copy()


def age_group_label(age):
    if age < 18:
        return 'child'
    elif age >= 65:
        return 'senior'
    else:
        return 'adult'

t_age['age_group'] = t_age['Age'].apply(age_group_label)


total_passengers = len(t_age)


counts = t_age['age_group'].value_counts().reindex(['child', 'adult', 'senior']).fillna(0).astype(int)


proportions = ((counts / total_passengers) * 100).round(2)


age_group_df = pd.DataFrame({
    'AgeGroup': counts.index,
    'Count': counts.values,
    'PctOfPassengers': proportions.values.round(2)
})

print(f"Total passengers considered: {total_passengers}")
display(age_group_df)



plt.figure(figsize=(6,4))
plt.bar(proportions.index, proportions.values)

plt.title("Percentage of Passengers in Every Age Group")
plt.xlabel("Age Group")
plt.ylabel("Percentage (%)")

# Add text labels on bars
for i, val in enumerate(proportions.values):
    plt.text(i, val + 1, f"{val}%")

plt.ylim(0, max(proportions.values)+10)
plt.show()
```
